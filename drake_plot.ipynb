```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

drake = pd.read_csv('drake_discography.csv')
zesty_drake = pd.read_csv('zesty_drake_discography.csv')

not_zesty_drake = drake[~drake.isin(zesty_drake)].dropna()
not_zesty_drake.to_csv('not_zesty_drake_discography.csv', index=False)
df_nz = pd.read_csv('not_zesty_drake_discography.csv')

%matplotlib inline
sns.pairplot(df_nz)
```
